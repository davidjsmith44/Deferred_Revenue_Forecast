## Creating FX forward rate numpy array
The rows correspond to the document currency
The columns correspond to spot and forward rates

NOTE: Ignoring positive and negative shocks to the FX exchange rates

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
FX_rates = pd.read_excel('../data/Data_2019_P06/FX_data.xlsx', 'to_matlab')
```

```python
FX_rates
```

```python
# Change the vols to be percentages
FX_rates['VOL_3M'] = FX_rates['VOL_3M']/100
FX_rates['VOL_6M'] = FX_rates['VOL_6M']/100
FX_rates['VOL_9M'] = FX_rates['VOL_9M']/100
FX_rates['VOL_1Y'] = FX_rates['VOL_1Y']/100
FX_rates
```

```python
term = [.25, .5, .75, 1]
```

```python
# need to sort the tickers based on direct or indirect
to_invert = FX_rates['Ticker'].str.slice(0,3).str.contains('USD')
for index, row in FX_rates.iterrows():
     if to_invert.iloc[index] == True:
            FX_rates['Spot'].loc[index] = 1/row['Spot']
            FX_rates['FWD_3M'].loc[index]  = 1/row['FWD_3M']
            FX_rates['FWD_6M'].loc[index]  = 1/row['FWD_6M']
            FX_rates['FWD_9M'].loc[index]  = 1/row['FWD_9M']
            FX_rates['FWD_1Y'].loc[index]  = 1/row['FWD_1Y']
```

```python
FX_rates
```

```python
# Interpolating the forward rates
FX_rates['FWD_1M'] = FX_rates['Spot']* (2/3) + FX_rates['FWD_3M']*(1/3)
FX_rates['FWD_2M'] = FX_rates['Spot']* (1/3) + FX_rates['FWD_3M']*(2/3)

FX_rates['FWD_4M'] = FX_rates['FWD_3M']* (2/3) + FX_rates['FWD_6M']*(1/3)
FX_rates['FWD_5M'] = FX_rates['FWD_3M']* (1/3) + FX_rates['FWD_6M']*(2/3)

FX_rates['FWD_7M'] = FX_rates['FWD_6M']* (2/3) + FX_rates['FWD_9M']*(1/3)
FX_rates['FWD_8M'] = FX_rates['FWD_6M']* (1/3) + FX_rates['FWD_9M']*(2/3)

FX_rates['FWD_10M'] = FX_rates['FWD_9M']* (2/3) + FX_rates['FWD_1Y']*(1/3)
FX_rates['FWD_11M'] = FX_rates['FWD_9M']* (1/3) + FX_rates['FWD_1Y']*(2/3)

FX_rates
```
