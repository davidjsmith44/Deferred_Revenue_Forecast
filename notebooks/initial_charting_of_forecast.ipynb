# Working on what we need to see from a graphing perspective

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

import pickle
#from math import ceil
#from sklearn.linear_model import LinearRegression
#from scipy.interpolate import interp1d, griddata
```

```python
model_dict = pickle.load( open( "../data/processed/initial_forecast.p", "rb" ) )
```

```python
model_dict.keys()
```

```python
df_fcst = model_dict['forecast']
df_fcst.head(10)
```

```python
df_hist = model_dict['billings']
df_hist.head(10)
```

```python
this_curr = 'GBP'
this_BU = 'Creative'
this_slice = df_hist[(df_hist['curr']==this_curr)&
                    (df_hist['BU']==this_BU)]


fig, axs = plt.subplots(6,1, figsize=(10,10))

axs[0].plot(this_slice['period'], this_slice['deferred_3Y_DC'])
axs[0].set_title('Three year deferred billings')

axs[1].plot(this_slice['period'], this_slice['deferred_2Y_DC'])
axs[1].set_title('Two year deferred billings')

axs[2].plot(this_slice['period'], this_slice['deferred_1Y_DC'])
axs[2].set_title('1 year deferred billings')

plt.tight_layout()
```
